### 2.2 - Line and Step Charts : Python

```python
# Load libraries. Pandas for loading data and plots. Matplotlib for plots. Numpy for the one rounding portion of my pie chart.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import Excel files into dataframe
worldpop = pd.read_excel('world-population.xlsm')
```

```python
worldpop.head()
```

```python
# Matplotlib basic plot for line chart
plt.figure(figsize=(10, 5))

plt.plot(worldpop['Year'], worldpop['Population'])
plt.title('World Population Growth 1960-2009', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population (in billions)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Matplotlib step plot. I really like intuitive nature of matplotlib.
plt.figure(figsize=(10, 5))

plt.step(worldpop['Year'], worldpop['Population'])
plt.title('World Population Growth 1960-2009', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population (in billions)', fontsize=14)
plt.grid(True)
plt.show()
```
